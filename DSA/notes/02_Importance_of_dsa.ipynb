#### DSA in Efficient Problem-Solving
DSA knowledge is fundamental to efficient problem-solving for several key reasons:

##### 1. Resource Optimization
DSA helps create solutions that minimize:
- Execution time (speed)
- Memory usage
- Network bandwidth
- Database access


```python
import time

def is_prime_inefficient(n):
    """Check if a number is prime by testing all numbers from 2 to n-1"""
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# More efficient approach: Only check divisors up to square root of n
def is_prime_efficient(n):
    """Check if a number is prime by testing numbers from 2 to sqrt(n)"""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Test with a large prime number
test_num = 7919  # A prime number

# Time the inefficient approach
start = time.time()
result1 = is_prime_inefficient(test_num)
end = time.time()
inefficient_time = end - start

# Time the efficient approach
start = time.time()
result2 = is_prime_efficient(test_num)
end = time.time()
efficient_time = end - start

print(f"Is {test_num} prime? {result1}")
print(f"Inefficient approach time: {inefficient_time:.6f} seconds")
print(f"Efficient approach time: {efficient_time:.6f} seconds")
print(f"Improvement factor: {inefficient_time/efficient_time:.1f}x faster")
```

##### 2. Scalability
Well-designed algorithms scale better as input sizes grow:


```python
# Demonstrating how different algorithms scale with input size
import matplotlib.pyplot as plt
import numpy as np

# Generate input sizes
input_sizes = np.arange(1, 21)

# Different algorithmic complexity functions
constant = np.ones_like(input_sizes)  # O(1)
logarithmic = np.log2(input_sizes)  # O(log n)
linear = input_sizes  # O(n)
linearithmic = input_sizes * np.log2(input_sizes)  # O(n log n)
quadratic = input_sizes**2  # O(n²)
cubic = input_sizes**3  # O(n³)
exponential = 2**input_sizes  # O(2^n)

# Plot the scaling behavior
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, constant, 'o-', label='O(1) - Constant')
plt.plot(input_sizes, logarithmic, 's-', label='O(log n) - Logarithmic')
plt.plot(input_sizes, linear, '^-', label='O(n) - Linear')
plt.plot(input_sizes, linearithmic, 'D-', label='O(n log n) - Linearithmic')
plt.plot(input_sizes, quadratic, '*-', label='O(n²) - Quadratic')
plt.plot(input_sizes, cubic, 'x-', label='O(n³) - Cubic')

# Plot exponential separately to avoid scale issues
plt.figure(figsize=(10, 6))
plt.semilogy(input_sizes[:10], exponential[:10], 'o-', label='O(2^n) - Exponential', color='red')
plt.grid(True)
plt.title('Exponential Growth')
plt.xlabel('Input Size (n)')
plt.ylabel('Operations (log scale)')
plt.legend()

# Show first plot again
plt.figure(1)
plt.grid(True)
plt.title('Algorithm Scaling Behavior')
plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.legend()
plt.tight_layout()
```

##### 3. Problem Decomposition
DSA teaches structured approaches to break complex problems into manageable parts:
- Divide and conquer
- Dynamic programming
- Greedy approaches
- Iterative refinement


```python
# Example: Using recursion to solve problems through decomposition
def tower_of_hanoi(n, source, auxiliary, target):
    """
    Solve the Tower of Hanoi puzzle
    n: number of disks
    source: source rod
    auxiliary: auxiliary rod
    target: target rod
    """
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    
    # Move n-1 disks from source to auxiliary using target as temporary
    tower_of_hanoi(n-1, source, target, auxiliary)
    
    # Move the largest disk from source to target
    print(f"Move disk {n} from {source} to {target}")
    
    # Move n-1 disks from auxiliary to target using source as temporary
    tower_of_hanoi(n-1, auxiliary, source, target)

# Solve Tower of Hanoi with 3 disks
print("Tower of Hanoi solution for 3 disks:")
tower_of_hanoi(3, 'A', 'B', 'C')
```

```python
# Sample interview problem: Two Sum
# (Find two numbers in an array that add up to a target)

def two_sum_naive(nums, target):
    """
    Find two numbers that add up to target
    Brute force approach - O(n²) time
    """
    n = len(nums)
    for i in range(n):
        for j in range(i+1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return None
```

```python
def two_sum_optimized(nums, target):
    """
    Find two numbers that add up to target
    Optimized approach using hash map - O(n) time
    """
    # Map of value -> index
    num_map = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    
    return None

# Test the functions
test_array = [2, 7, 11, 15]
test_target = 9

print(f"Two Sum Naive: {two_sum_naive(test_array, test_target)}")
print(f"Two Sum Optimized: {two_sum_optimized(test_array, test_target)}")
```
